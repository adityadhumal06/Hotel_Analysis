# Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
```

```python
cancelled_perc = round(cancelled_perc,2)
print(cancelled_perc)

plt.figure(figsize=(10,8))
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Reservation Status is different is hotes')
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

```python
Resort_Hotel = df[df['hotel']=='Resort Hotel']
Resort_Hotel = round(Resort_Hotel['is_canceled'].value_counts(normalize=True),2)
Resort_Hotel
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel = round(city_hotel['is_canceled'].value_counts(normalize=True),2)
city_hotel
```

```python
Resort_Hotel = Resort_Hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```
